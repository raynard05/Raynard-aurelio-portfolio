```python
# Data
data_jam = [2, 3, 5, 6, 8]
data_nilai = [70, 75, 85, 90, 95]
```

```python
banyak_data = len(data_jam)
```

```python
mean_data_jam = sum(data_jam) / banyak_data
mean_data_nilai = sum(data_nilai) / banyak_data
```

```python
numerator = 0.0
denominator = 0.0

for i in range(banyak_data):
    dx = data_jam[i] - mean_data_jam
    dy = data_nilai[i] - mean_data_nilai
    numerator += dx * dy
    denominator += dx * dx

beta1 = numerator / denominator
```

```python
beta0 = mean_data_nilai - beta1 * mean_data_jam
```

```python
print(f"mean_data_jam = {mean_data_jam:.4f}")
print(f"mean_data_nilai = {mean_data_nilai:.4f}")

print(f"\nbeta1 (slope) = {beta1:.4f}")
print(f"beta0 (intercept) = {beta0:.4f}")

print(f"\nModel regresi: ŷ = {beta0:.4f} + {beta1:.4f}x")
```

```python
x_input = float(input("\nMasukkan nilai x untuk prediksi model regresi linier: "))
y_pred = beta0 + beta1 * x_input

print(f"Hasil prediksi untuk x = {x_input} adalah y = {y_pred:.4f}")
```

```python
SSE = 0
y_pred = []

for xi, yi in zip(data_jam, data_nilai):
    y_hat = beta0 + beta1 * xi
    y_pred.append(y_hat)
    SSE += (yi - y_hat)**2

print("SSE =", round(SSE, 5))
```

```python
SST = 0
for yi in data_nilai:
    SST += (yi - mean_data_nilai)**2

print("SST =", round(SST, 5))
```

```python
SSR = SST - SSE
print("SSR =", round(SSR, 5))
```

```python
R2 = 1 - (SSE / SST)
print("R^2 =", round(R2, 4))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Garis regresi
x_line = np.linspace(min(data_jam), max(data_jam), 100)
y_line = beta0 + beta1 * x_line

plt.figure(figsize=(5,5))

# Scatter data asli
plt.scatter(data_jam, data_nilai, label="Data Asli")

# Garis regresi
plt.plot(x_line, y_line, label="Garis Regresi")

# Titik prediksi
plt.scatter([data_jam], [y_pred], s=100, marker='X', label=f"Prediksi (x={data_jam})")

plt.xlabel("x (jam)")
plt.ylabel("y (nilai)")
plt.title("Regresi Linier + Titik Prediksi")
plt.legend()
plt.grid(True)
plt.show()
```
